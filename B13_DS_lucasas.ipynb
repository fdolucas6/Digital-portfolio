CONTOUR PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 20)
ylist = np.linspace(-3.0, 3.0, 20)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)

fig = plt.figure()
ax1= fig.add_subplot(111)

cp = ax1.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax1.set_title('Filled Contours Plot')
ax1.set_xlabel('x (cm)')
ax1.set_ylabel('y (cm)')

ax2 = fig.add_subplot(212)
ax2.scatter(X, Y,c=Z)

plt.show()
```

QUIVER PLOT

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points
xlist = np.linspace(-3.0, 3.0, 20)
ylist = np.linspace(-3.0, 3.0, 20)
X, Y = np.meshgrid(xlist, ylist)

# Define vector components
U = -1 - X**2 + Y  # x-component of the vector
V = 1 + X - Y**2   # y-component of the vector

# Create the quiver plot
fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V, color='r')  # Plot arrows with red color

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Quiver Plot Example')

# Set limits for better visualization
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

# Show the plot
plt.grid()
plt.show()
```
